# 5. Use Map, Filter, Reduce and Lambda Functions on List using Numpy

```python
import numpy as np
from functools import reduce as r
a=np.array([2,5,8,6,3])
m=list(map(lambda x:x**2,a))
f=list(filter(lambda x:x>4,a))
re=r(lambda x,y:x+y,a)
print(a)
print("map",m)
print("filter",f)
print("reduce",re)
```

# 7. Grouping using Pandas on a Dataframe

```python
import pandas as pd
import numpy as np
a=pd.DataFrame({
            'Course':["PY","JV","DBMS","MMA","MMA"],
            'Fee':[300,600,21,350,67],
            'Complexity':[100,56,32,10,67]})
print(a)
c=a.groupby('Course').agg({'Fee':'min'})
print(c)
```

```python
import pandas as p
t={
'Course':["PY","JV","DBMS","MMA","MMA"],
'Fee':[300,600,21,350,67],
'Complexity':[100,56,32,10,67]
}
d=p.DataFrame(t)
print(d)
c=d.groupby('Course').agg({'Fee':'min'})
print("\n",c)
```

# 8. Pivot and melt functions using Pandas.

```python
import pandas as p
t={
'Course':["PY","JV","DBMS","MMA","MMA"],
'Fee':[300,600,21,350,67],
'Complexity':[100,56,32,10,67]
}
d=p.DataFrame(t)
print(d)
print("\n",d.pivot(columns='Course',values='Complexity'))
print("\n",d.melt())
```

# 9. Use Map, Filter and Reduce, Lambda functions using Pandas [Data Frame]


```python
from functools import reduce as r
import pandas as pd
data={
    'Numbers':[5,2,6,7,1,3,4],
    'letter':['A','B','C','D','E','F','G']}
df=pd.DataFrame(data)
sq=df['Numbers'].map(lambda x: x**2)
ev=list(filter(lambda x: x % 2 == 0, df['Numbers']))
po = r(lambda x, y: x * y, df['Numbers'])
print("Dataframe:\n",df)
print("\nMap for Squaring:\n",sq)
print("\nfilter for even numbers:\n", ev)
print("\nReduce for product:\n", po)
```

# 10. Time series using Pandas (resample, shift operations)

```python
import numpy as np
import pandas as pd
d=p.DataFrame(
{"date":pd.date_range(start="2023-09-07",periods=5,freq="D"),
 "temp":np.random.randint(18,30,size=5)}
)
print("d",d)
d["f"]=d["temp"].shift(1)
print("Shift:\n",d)
dfw=d.resample("ME",on="date").mean()
print("\nResampling:\n",dfw)
```

# 13. Visualization of Iris-dataset using Scatter Plot

```python
import pandas as pd
from matplotlib import pyplot as plt
t = pd.read_csv("D:\\kaveri\\IRIS.csv")
species_colors = {
    'Iris-setosa': 'k',
    'Iris-versicolor': 'g',
    'Iris-virginica': 'r'}
for spe, color in species_colors.items():
    sl = t[t['species'] == spe]['sepal_length']
    sw = t[t['species'] == spe]['sepal_width']
    plt.scatter(sl, sw, color=color, label=spe)
plt.legend()
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Width and Length for Iris Species')
plt.show()
```
